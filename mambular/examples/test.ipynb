```python
import torch
import torch.nn.functional as F
```

```python
def deal_preds(preds):
    preds_1 = F.softmax(preds[:,:3],dim=1)
    preds_1_index = torch.argmax(preds_1,dim=1)
    preds_2 = F.softmax(preds[:,3:],dim=1)
    preds_2_index = torch.argmax(preds_2,dim=1)
    preds_combined = torch.stack([preds_1_index,preds_2_index],dim=1)
    return preds_combined.float()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# 模拟数据：每个类别的正负样本数量
data = {
    '类别': ['低体重儿', '巨大儿', '早产', '胎儿死亡', '畸形'],
    '正类': [1472642, 1497259, 1444140, 1547027, 1535173],
    '负类': [82313, 57696, 110815, 7928, 19782]
}

# 创建DataFrame
df = pd.DataFrame(data)

# 设置柱的宽度
bar_width = 0.35

# 创建索引
index = np.arange(len(df['类别']))

# 绘制正负类样本的柱状图
plt.bar(index, df['正类'], bar_width, label='正类')
plt.bar(index + bar_width, df['负类'], bar_width, label='负类', color='orange')

# 添加标签和标题
plt.xlabel('类别')
plt.ylabel('计数')
plt.title('各类的正负样本计数')

# 设置x轴的刻度和标签
plt.xticks(index + bar_width / 2, df['类别'], rotation=45)
plt.tick_params(axis='x', direction='in', length=5)
plt.tick_params(axis='y', direction='in', length=5)
# 添加图例
plt.legend()

# 显示图像
plt.tight_layout()
plt.show()
```
